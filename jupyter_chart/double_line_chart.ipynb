```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
```

```python
# initialize data
qps_recall_xchg_nbr = pd.DataFrame(
  [
    [2133.04, 41.26], 
    [2097.99, 41.86],
    [2071.08, 42.42],
    [2042.78, 42.98],
    [2015.49, 43.51],
    [1989.43, 44.02]
  ], columns = ['qps', 'recall'])

qps_recall_original = pd.DataFrame(
[
   [2134.42, 39.96], 
   [2107.07, 40.63],  
   [2073.97, 41.24],  
   [2043.42, 41.84],  
   [2016.03, 42.00],  
   [1992.96, 42.96],  
], columns = ['qps', 'recall'])
```

```python
qps_recall_original['recall']
```

```python
qps_recall_xchg_nbr['recall']
```

```python
fig, ax = plt.subplots()
# fig = plt.figure(1)
# ax = fig.add_subplot(1,2,1)
# legend_ax = fig.add_subplot(1,2,2)

for index,row in qps_recall_original.iterrows():
    if index == 4:
        plt.text(row['qps']-16,row['recall']-1.4,str(row['qps'])+",\n"+str(row['recall']),color='b')
    else:
        plt.text(row['qps']-16,row['recall']-1.7,str(row['qps'])+",\n"+str(row['recall']),color='b')

for index, row in qps_recall_xchg_nbr.iterrows():
    plt.text(row['qps'],row['recall']+0.3,(row['qps'], row['recall']),color='g')


original = ax.plot(qps_recall_original['qps'], qps_recall_original['recall'], color=sns.xkcd_rgb["blue"], marker='*', linestyle='-', linewidth=1, markersize=6)
xchg_nbr = ax.plot(qps_recall_xchg_nbr['qps'], qps_recall_xchg_nbr['recall'], color=sns.xkcd_rgb["green"], marker='x', linestyle='-', linewidth=1, markersize=6)

ax.set_xlim(1960,2200)
ax.set_ylim(35,45)
ax.set_ylabel("QPS",fontsize=11)



#ax.set_xticks(x1_list)
#ax.set_xticklabels(recall_qps['recall'])
ax.set_xlabel('#qps',fontsize=11)

# 坐标轴标签设置
ax.set_title('QPS vs recall@10 w/ system cache',fontsize = 14)

# x轴标签旋转
#ax.set_xticklabels(ax.get_xticklabels(),rotation = 25)

# # 网格设置
# plt.grid('off')


# 图例设置
plt.legend(
    loc=2, bbox_to_anchor=(1.15,1.0), borderaxespad=0.,
    handles = [
        Line2D([0], [0], color=xchg_nbr[0].get_color(),marker=xchg_nbr[0].get_marker(), lw=xchg_nbr[0].get_linewidth(), label='xchg_neighbors'),
        Line2D([0], [0], color=original[0].get_color(),marker=original[0].get_marker(), lw=original[0].get_linewidth(), label='original'),
    ])

# set legend loc
# ax.legend(loc=2, bbox_to_anchor=(1.05,1.0), borderaxespad=0.)
plt.show()
```
